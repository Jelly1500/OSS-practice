### 2. 코드블럭코드("```"이용)
- - -
```
def hanoi(n, a, b, c):
    if n == 1:
        print(a, "->", b)
    hanoi(n-1, a, c, b)
    print(a, "->" b)
    hanoi(n-1, c, b, a)
```
